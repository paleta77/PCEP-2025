```python
# 1. wypisanie "Hello World!" 5 razy przy użyciu pętli for (for _)
for _ in range(5):
    print("hello world")
```

```python
# 2. odliczanie od 1 do 10 (for zmienna, range(1,11))
for number in range(1, 11):
    print(number)
```

```python
# 3. wypisanie liter ze stringa za pomocą pętli for (for in)
text = "qweasdzxc"
for number in range(0,5):
    print(text[number])
```

```python
# 4. wypisanie liter za pomocą indexu # (for i in range(len(my_list)))
text = "qweasdzxc"
for number in range(len(text)):
    print(f"{number}: " + text[number])
```

```python
# 5. odliczanie od 10 do 1 przy użyciu pętli while
countdown = 10
while countdown > 0:
    print(countdown)
    countdown -= 1
```

```python
# 6. proste menu z użyciem pętli while true
print("wybierz opcje:")
print("1. opcja")
print("2. opcja druga")
print("3. zakończ program")
while True:
    option = int(input("Wybierz opcje: "))
    if option == 1:
        pass
    elif option == 2:
        print("wybrano opcje drugą")
    elif option == 3:
        break
    else:
        print("podano zły wybór")
```

Ćwiczenie 2. 
Stwórz pętle while która odliczy 1 do 10 (odwrotnie niż w przykładzie)

```python
# 7. pass
for number in range(10):
    pass
```

```python
# 8. break
for number in range(10):
    print(number)
    if number == 5:
        break
```

```python
# 9. continue
for number in range(10):
    if number == 5:
        continue
    print(number)
```

Ćwiczenie 3.
Stwórz dowolną pętle która wyświetli tylko liczby nieparzyste od 3 do 12  z pominięciem 5 <3,12> (4)

```python
# 10. for else
for number in range(10):
    print(number)
else:
    print("koniec pętli")
```

```python
# 11. while else
number = 10
while number > 0:
    print(number)
    number -= 1
else:
    print("start")
```

```python
# 12. Pętle zagnieżdżone (wpisz 3 razy 123)
for _ in range(3):
    for number in range(1,4):
        print(number)
```
