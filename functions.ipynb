```python
# 7. funkcje z rekurencją
def silnia(n):
    if n == 0:
        return 1
    else:
        return n * silnia(n-1)
    
print(silnia(5))  # Output: 120
```

```python
# 9. variable shadowing (nonlocal)
def outer_function():
    x = 5
    def inner_function():
        nonlocal x
        x = 10
        print("Inner:", x)
    inner_function()
    print("Outer:", x)
outer_function()
```

```python
# 10. generatory
def generuj_liczby():
    for i in range():
        yield i * 2

for liczba in generuj_liczby():
    print(liczba)
```
