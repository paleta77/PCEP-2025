```python
# 1. tworzenie słownika 
car = {
    "brand": "volvo",
    "power": 150,
    "unit_of_power": "km",
    100: "test"
}
```

```python
# 2. wypisanie elementu ze słownika ([])
brand = car["brand"]
power = car["power"]
unit = car["unit_of_power"]
print(f"Samochód marki {brand} posiada moc {power} {unit}")
```

```python
# 3. wypisanie elementu ze słownika (get(klucz, domyślna_wartość))
print(car.get("brand"))
print(car.get("color", "N/A"))
```
