```python
# 1. tworzenie listy
list = [1, "number", 1.45]
for element in list:
    print(element)
```

```python
 # 2. sprawdzenie długości listy
 
list = [1, "number", 1.45]
print(len(list))
```

```python
# 3. wypisanie wybranych elementów listy (index i slicing <,))
list = [1, "number", 1.45]
print(list[1])
```

```python
 # 4. wypisanie zawartości listy za pomocą pętli for

list = [1, "number", 1.45]

for element in list:
    print(element)
    
print(list)

length = len(list)
while length >=0:
    print(list[length-1])
    length -= 1
```

```python
# 11. list comprehension (tworzenie listy za pomocą pętli w jednej linii) (podnieś każdą liczbę z listy do kwadratu)

numbers = [3, 20, 22, 1, 0, 10, 30]
powers = [ number ** 2 for number in numbers ]
print(powers)
```

```python
# 12. list comprehension z warunkiem (podnieś do kwadratu tylko liczby parzyste)
numbers = [3, 20, 22, 1, 0, 10, 30]
powers = [number ** 2 for number in numbers if number % 2 == 0 and number != 0]
print(powers)
```
