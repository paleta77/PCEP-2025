```python
# 13. różnica między mutable i immutable (int vs list)
lista = [3, 22, 1, 0, 34, 5, 7, 8]
lista2 = lista
lista2 += [99, 100]
print(lista)
print(id(lista))
print(id(lista2))

lista3 = lista.copy()
print(id(lista3))

lista4 = list(lista)
print(id(lista4))

```

```python
# 14. różnica między mutable i immutable (int vs list)

x = 10
print("x " + str(id(x)))
y = x
print("y " + str(id(y)))
y += 5
print("x " + str(id(x)))
print("y " + str(id(y)))
print(x)
print(y)
```
